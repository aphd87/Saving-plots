```python
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
body_data = pd.read_csv("weights_and_heights.csv")

#Split data frames
male = body_data[body_data["Gender"] == "Male"]
female = body_data[body_data["Gender"] == "Female"]

male_sample = male.sample(200, random_state = 42) #Random 200 sample, 42 means multiple people should get same randomly
#We can pick patient #705 
patient = male_sample.loc[[705]]

#We are adding min and median lines here
median_weight = male_sample["Weight"].median()
median_height = male_sample["Height"].median()
min_weight = male_sample["Weight"].min()
min_height = male_sample["Height"].min()
```

```python
#To add text, use annotate method
#Let's highlight the data point here
plt.style.use("seaborn-poster")
plt.scatter(male_sample["Weight"], male_sample["Height"], color = "blue", s = 700, alpha = 0.5) #s= size of data points, alpha = transparency of data points
plt.scatter(patient["Weight"], patient["Height"], color = "pink", s = 700, alpha = 1.0, edgecolor = "red", linewidth = 5) #New data point
plt.scatter(patient["Weight"], patient["Height"], marker = "x", color = "red", s = 250, alpha = 1.0, edgecolor = "pink", linewidth = 5) #New data point
plt.title("Weight vs Height for Males")
plt.xlabel("Weight (lbs)")
plt.ylabel("Height (in)")
plt.axvline(x = median_weight, color = "black", linestyle = "--" ) #Median_weight
plt.axhline(y = median_height, color = "black", linestyle = "--" ) #Median_height

plt.annotate(text = "Median Weight", xy = (190, 60), fontsize=16) #190 and #60 are about right here #This appears at bottom

plt.tight_layout()
plt.show()
```

```python
#To add text, use annotate method
#Let's highlight the data point here
plt.style.use("seaborn-poster")
plt.scatter(male_sample["Weight"], male_sample["Height"], color = "blue", s = 700, alpha = 0.5) #s= size of data points, alpha = transparency of data points
plt.scatter(patient["Weight"], patient["Height"], color = "pink", s = 700, alpha = 1.0, edgecolor = "red", linewidth = 5) #New data point
#plt.scatter(patient["Weight"], patient["Height"], marker = "x", color = "red", s = 250, alpha = 1.0, edgecolor = "pink", linewidth = 5) #New data point
plt.title("Weight vs Height for Males")
plt.xlabel("Weight (lbs)")
plt.ylabel("Height (in)")
plt.axvline(x = median_weight, color = "black", linestyle = "--" ) #Median_weight
plt.axhline(y = median_height, color = "black", linestyle = "--" ) #Median_height

plt.annotate(text=f"Median Weight: {round(median_weight)} lbs", #This is where we place a median weight call out
             xy=(median_weight, min_height),
             xytext=(10, -10),
             textcoords="offset pixels",
             fontsize=16)

plt.annotate(text=f"Median Height: {round(median_height)} lbs", #This is where we place a median weight call out
             xy=(min_weight, median_height),
             xytext=(-10, 10),
             textcoords="offset pixels",
             fontsize=16)

plt.annotate(text = "Patient 705",  
            xy = (patient["Weight"], patient["Height"]),
            xytext = (140,74),
            fontsize = 25,
            bbox = dict(boxstyle="round", #bounding box where text sits in
                       fc = "salmon", #fill color for bounding box
                       ec = "red"), #edge color of bounding box
            arrowprops = dict(arrowstyle = "wedge, tail_width=1.",
                             fc = "salmon",
                             ec = "red",
                             patchA = None, #We want arrow to sit up against box with no gap
                             connectionstyle = "arc3,rad=-0.1")) #Shape and curve of connector arrow
plt.tight_layout()
plt.savefig(fname = "exported_plot.png")#
plt.show()
```
